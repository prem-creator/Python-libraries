# Seaborn

* Seaborn is a python library for data visualization build on top of the matplotlib python library

* Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.

```python
# importing libraries
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
```

## Lineplot

```python
days = list(range(1, 16))
temperature = [36.5 , 38 , 37 , 39 , 40 , 43 , 45 , 46.4 , 47 , 48 , 49 , 46.5 , 49 , 45 , 43 ]
```

```python
sns.lineplot(x = days ,y = temperature)
```

```python
df = pd.DataFrame({"days" : days , "temperature" : temperature})
```

```python
df.head()
```

```python
sns.lineplot(x = "days" , y = "temperature" , data = df)
```

```python
df_tip = sns.load_dataset("tips")
```

```python
df_tip
```

```python
plt.figure(figsize = (16,9))
sns.set(style = "darkgrid")
sns.lineplot(x= "tip" ,y = "total_bill", data = df_tip ,hue = "sex",  style = "sex" ,
            markers = (["o" , "<"]))
plt.ylabel("total bill", fontsize = 20)
plt.xlabel("tip", fontsize = 20)
plt.title("line_plot" , fontsize = 15 , color = "g")
plt.show()
```

## Histogram

```python
#bins = [1,5,10 ,15,20]
sns.distplot(df_tip["total_bill"] , kde = True , color = "r" , vertical = False , axlabel = "total_bill" ,
             label = "total_bill")
plt.legend()
plt.title("Histogram", fontsize = 18)
plt.ylabel("probability density" , fontsize = 15)
#plt.xticks(bins)
plt.show()
```

## Scatter plot

```python
df_1 = sns.load_dataset("titanic")
df_1
```

```python
plt.figure(figsize = (16,9))
sns.scatterplot(x = "age" , y = "fare" , data = df_1 , hue = "sex" , style = "who" ,alpha = 0.2)
```

## heatmap

## jointplot

```python
sns.jointplot(x='tip',y='total_bill',data=df_tip,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df_tip,kind='reg')
```

```python
sns.pairplot(df_tip,hue='sex') # multivariate analysis
```

## Count plot

```python
sns.countplot(x='sex',data=df_tip)
```

## Bar plot

```python
sns.barplot(x='sex' , y='total_bill',data=df_tip)
```

## boxplot

```python
sns.boxplot('smoker','total_bill', data=df_tip)
```

## Violinplot

```python
sns.violinplot(x="total_bill", y="day", data=df_tip,palette='rainbow')
```
