### Lineplot

```python
import matplotlib.pyplot as plt
```

```python
days = list(range(1,16))
```

```python
delhi_temperature = [25 , 24 , 23 , 24.5 , 20.2 , 24 , 23.5 , 22 , 28 , 30 , 32.1 , 21 , 20 , 24 , 25.5]
```

```python
mumbai_temperature = [20 , 18 , 24,25,22.2 , 23.5 , 20.5 , 18.7 , 17 , 19 , 21.5 , 22 , 23 , 20.4 , 22.3]
```

```python
plt.plot(days,delhi_temperature)
plt.show()
```

```python
plt.plot(days,delhi_temperature)

plt.title(" weather forcast")
plt.show()
```

```python
plt.plot(days,delhi_temperature)
plt.title(" weather forcast")
plt.xlabel("days")
plt.ylabel("temperature")
plt.show()
```

```python
plt.plot(days,delhi_temperature)
plt.title(" weather forcast")
plt.axis([0,15,0,40])
plt.xlabel("days" , color = "blue")
plt.ylabel("temperature" , color = "blue")
plt.legend( "temperature_line",loc =2)
plt.show()
```

```python
from matplotlib import style
```

```python
plt.plot(days,delhi_temperature,color= "g",marker= "<", linestyle = "--", linewidth = 2 , markersize = 10 ,
         label= "delhi_temperature")
plt.plot(days,mumbai_temperature,color= "y",marker= "o", linestyle = "--", linewidth = 2 , markersize = 10 ,
         label= "mumbai_temperature")
plt.title(" weather forcast",color= "r")
plt.axis([0,15,0,40])
plt.xlabel("days" , color= "blue")
plt.ylabel("temperature", color= "blue")
plt.legend(loc =1)
style.use("ggplot")
plt.grid(color = "black", linestyle = "--", linewidth = 1)
plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
marks_maths_students = np.random.randint(42,85,(100))
```

```python
marks_maths_students 
```

```python
plt.hist(marks_maths_students )
plt.title("Marks")
plt.xlabel("Marks of student")
plt.ylabel("number of students")
plt.show()
```

```python
bins = [40 , 50, 60, 70, 80, 90, 100]
```

```python
plt.hist(marks_maths_students ,bins)
plt.title("Marks")
plt.xlabel("marks of student")
plt.ylabel("number of students")
plt.show()
```

```python
plt.hist(marks_maths_students , rwidth=0.8, histtype= "bar",
         orientation='vertical',color = "orange",label= "marks of maths")
plt.title("Marks" , fontsize = 18)
plt.xlabel("marks of student" , fontsize = 14)
plt.ylabel("number of student" , fontsize = 14)
plt.legend()
plt.show()
```

```python
marks_chemistry_students = np.random.randint(42,90,(100))
```

```python
from matplotlib import style
```

```python
plt.figure(figsize = (14,6))
plt.hist([marks_maths_students,marks_chemistry_students] , rwidth=0.8, histtype= "bar",    align='mid',
         orientation='vertical',color = ["orange", "m"],label= ["marks of maths", "marks of chemistry"])
plt.title("Marks")
plt.xlabel("marks of student")
plt.ylabel("number of student")
plt.legend()
style.use("ggplot")
plt.grid(color = "black", linestyle = "-", linewidth = 1)
plt.show()
```

### Barchart

```python
import  matplotlib.pyplot as plt
import numpy as np
```

```python
classes = ["data analysis", "Auditing", "bussiness environment", "Management Accounting", "IT in bussiness"]
Batch_1 = ["30", "60", "35", "30", "50"]
Batch_2 = ["35", "65", "30", "35", "45"]
```

```python
plt.bar(classes, Batch_1)
```

```python
plt.barh(classes, Batch_1)
```

```python
plt.bar(classes, Batch_1, width =0.8
        , align='edge', color ="y",edgecolor = "green", linewidth=6, alpha=1, linestyle="-",label="batch_1")
plt.legend()
plt.show()
```

```python
classes_index = np.arange(len(classes))
width = 0.2
```

```python
plt.barh(classes, Batch_1 , width
        , align='edge', color ="y",edgecolor = "green", linewidth=6,alpha=1, linestyle="--",label="batch_1")
plt.barh(classes_index+width, Batch_2,color ="orange",edgecolor = "blue",linewidth=5,label="batch_2")
plt.xticks(classes_index+width, classes, rotation = 50)
plt.legend()
plt.show()
```

### scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("SalaryGender.csv")
```

```python
df.head()
```

```python
x = df["Salary"]
y = df["Age"]
```

```python
plt.figure(figsize = (14,6))
plt.scatter(x,y , c="r", marker="*", s=500 , alpha=1 , linewidth=2 ,edgecolor= "y")
plt.title("salary")
plt.xlabel("salary")
plt.ylabel("age")
plt.show()
```

### Pie chart

```python
import matplotlib.pyplot as plt
```

```python
classes = ["data analysis", "Auditing", "bussiness environment", "Management Accounting", "IT in bussiness"]
Batch_1 = ["30", "60", "35", "30", "50"]
Batch_2 = ["35", "65", "30", "35", "45"]
```

```python
explode = [0,0,0.1,0,0.2]
colors = ["r","y","g","b","pink"]
plt.pie(Batch_1 , labels = classes,explode=explode,colors=colors ,
        autopct="%0.2f%%",shadow=True ,  radius=1.5,startangle=90, textprops={"fontsize":15},)
plt.show()
```
