### DataFrame

1. **Handling Missing  values( isnull()  ,  dropna())**
2. **fillna()**
  * fillna fills the nan values with some other  values
3. **replace()**
  * replaces the values of dataframe with other values
* **interpolate()**
  * Pandas interpolate() function is basically used to fill nan values in dataframe or series
* **loc and iloc**
  * we can access the rows and columns by thier label and number
* **groupby()**
  * it is used to split data into groups based on some criteria
* **Merging()**
  * for combining data on common columns or indice
* **Concat()**
  *  for combining DataFrames across rows or columns
* **join()**
  * for combining data on a key column or an index
* **Append()**
  * it is used to append rows of other dataframe to the end of the given dataframe , returning a new dataframe
* **Pivot table**
  * levels in the pivot table will be sorted in multi-index objects (hierarchical indexes) on the index and columns of the result dataframe
* **Melt()**
  * this function is used to transform or reshape data

### Merge()

```python
df4 = pd.DataFrame({"ID" : [1,2,3,4] , "class" : [9,110,1,12]})
```

```python
df5 = pd.DataFrame({"ID" : [1,2,3,5] , "Name" : ["a" , "b" , "c" , "d"]})
```

```python
pd.merge(df4,df5 , on = "ID")
```

```python
pd.merge(df4 , df5 , on ="ID" , how = "inner" ) # use intersection of keys from both frames
```

```python
pd.merge(df4 , df5 , on ="ID" , how = "left" ) # use only keys from left frame
```

```python
pd.merge(df4 , df5 , on ="ID" , how = "right" ) # use only keys from right frame
```

```python
pd.merge(df4 , df5 , on ="ID" , how = "outer" ) # use union of keys from both frames 
```

### Concat()

```python
df6 =pd.DataFrame(({"ID" : [1,2,3,4],
       "Name" : ["a", "b","c", "d"],
       "Class" : [5,6,7,8]}))
```

```python
df7 =pd.DataFrame({"ID" : [1,2,3,4],
       "Name" : ["e", "f","g", "h"],
       "Class" : [9,10,11,12]})
```

```python
pd.concat([df6,df7])
```

```python
pd.concat([df6,df7] , axis = 1)
```
